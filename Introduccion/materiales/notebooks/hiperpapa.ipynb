```python
import math
from matplotlib import pyplot as plt
from scipy.special import gamma
```

```python
def volumen_papa(n, r):
    return (math.pi ** (n / 2) * r ** n) / gamma(n / 2 + 1)
```

```python
r=3.0
dr=0.1

dimensiones = []
cocientes = []

for n in range(300):
    cociente = 100*(volumen_papa(n,r) - volumen_papa(n,r-dr))/volumen_papa(n,r)
    cocientes.append(cociente)
    dimensiones.append(n)
```

```python
plt.plot(dimensiones, cocientes)
plt.title("Volumen de capa / Volumen total\nHiperpapa de 3 cm y capa de 1 mm")
plt.xlabel("Dimensionalidad")
plt.ylabel("Porcentaje")
plt.savefig('hiperpapa.jpg', format='jpg', dpi=300)
plt.show()
```
